```python
rosters = pd.read_csv('AUDL_Rosters.csv')
```

```python
def GetSuggestion1(row):
    return GetSuggestion(row,1)
def GetSuggestion2(row):
    return GetSuggestion(row,2)   
def GetSuggestion3(row):
    return GetSuggestion(row,3)  
def GetSuggestion(row,no):
    roster = rosters[(rosters.Teamname==row['Teamname'])&(rosters.Tournament==row['Tournament'])].PlayerName.dropna().unique()
    matches = difflib.get_close_matches(row['Username'], roster,cutoff=.3)
    return matches[no-1] if len(matches) > no-1 else np.nan

upr['Suggestion1'] = upr.apply(GetSuggestion1,axis=1)
upr['Suggestion2'] = upr.apply(GetSuggestion2,axis=1)
upr['Suggestion3'] = upr.apply(GetSuggestion3,axis=1)
                                                                                                                                      
                                                                                                                                                        
                                   
```

```python
# upr[upr.PlayerName!=upr.Suggestion1]
upr['PlayerName'] = upr.PlayerName.fillna(upr.Suggestion1)
```
