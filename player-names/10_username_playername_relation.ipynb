```python
import glob
import difflib
import pandas as pd

from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

## Find Poor Matches

```python
similar('A Olsen','Anders Olsen')==similar('Anders Olsen','A Olsen')
```

```python
upr = pd.read_csv('09_username_playername_relation.csv',encoding = "ISO-8859-1")
upr.head()
```

```python
upr['MatchScore'] = upr.fillna('').apply(lambda x : similar(x.Username,x.PlayerName),axis=1)
```

```python
upr.to_csv('10_username_playername_relation.csv',index=False)
```
